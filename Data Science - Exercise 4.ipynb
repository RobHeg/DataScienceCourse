1. Write a NumPy program to get the indices of the sorted elements of a given array.



```python
#given this array
import numpy as np
score = np.array([103, 1502, 6230, 3, 1682, 5241, 4532])
score
```

```python
#sorting the array is possible with np.sort()
sortedScore = np.sort(score)
sortedScore
```

```python
#initialize empty array
sortedIndices = np.zeros(len(sortedScore), dtype=np.int64)

#find the indice that matches each element from sortedScore in score
for i in range(len(sortedScore)):
    sortedIndices[i] += np.where(score == sortedScore[i])

sortedIndices
```

```python
#turns out there is already a function for all of that...
argSort = score.argsort()
argSort
```

```python
#Load the countries.csv directly via URL import into your panda data frame!
import pandas as pd
df = pd.read_csv('https://github.com/WHPAN0108/BHT-DataScience-S23/blob/main/python-DS/country.csv?raw=true')
df
```

```python
#Show all the rows of countries that have the EURO
eur_df = df[df['Currency'] == 'EUR'] #getting rid of the first df just returns true or false values
eur_df
```

```python
df.loc[df['Currency'] == 'EUR']
```

```python
#Show only name and Currency in a new data frame
namcur_df = df[['Name', 'Currency']]
namcur_df
```

```python
df_copy = df.copy() #because the previous one technically still has a reference towards the old df and is not completely new
namcur2_df = df_copy[['Name', 'Currency']]
namcur2_df
```

```python
#Show only the rows/countries that have more than 2000 GDP (it is in Milliarden USD Bruttoinlandsprodukt)
df[df['GDP'] > 2000]
```

```python
#Select all countries where with inhabitants between 50 and 150 Mio
df[(50000000 < df['Population']) & (df['Population'] < 150000000)]
#I assume that 50 mio and 150 mio exactly are not in between; else use <=.
```

```python
#Calculate the GDP average (ignore the missing value)
df['GDP'].mean() #the skipna parameter of mean is default true, meaning nans are automatically ignored
```

```python
#Calculate the GDP average (missing value treated as 0)

#replace nans with 0
nan0_df = df.fillna(0)
nan0_df
```

```python
nan0_df['GDP'].mean()
#lower as expected
```

```python
#Calculate the population density (population/area) of all countries and add as new column
df['Density'] = df['Population'] / df['Area'] #so intuitive!
df
```

```python
#Sort by country name alphabetically
df.sort_values('Name')
```

```python
#Create a new data frame from the original where the area is changed: all countries with > 1000000 get BIG 
#and <= 1000000 get SMALL in the cell replaced!
area_df = df.copy()                                     #crucial to not just say area_df = df here.
area_df.loc[df['Area'] > 1000000, 'Area'] = 'BIG'
area_df.loc[df['Area'] <= 1000000, 'Area'] = 'SMALL'    #need to use df['Area'] instead of area_df['Area'] like here
                                                        #because else we are comparing BIG with a number (<= 1000000)
area_df
```
