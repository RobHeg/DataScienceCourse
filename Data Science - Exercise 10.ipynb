# Data preparation

```python
from sklearn.cluster import KMeans
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
data = df.select_dtypes(include=['float64', 'int64'])
```

```python
# outlier analysis - we've done this before
attributes = df.select_dtypes(include=['float64', 'int64']).columns
plt.rcParams['figure.figsize'] = [10,8]
sns.boxplot(data=df[attributes], orient="v", palette="Set2", whis=1.5, saturation=1, width=0.7)
plt.title("Outliers Variable Distribution", fontsize=14, fontweight='bold')
plt.ylabel("Range", fontweight='bold')
plt.xlabel("Attributes", fontweight='bold')
```

```python
'''This outlier removal removes Singapore for exports, Qatar for income and Nigeria (potentially a country to help) for inflation'''
# Remove statistical outliers for each attribute and show which outliers were removed
for attribute in attributes:
    Q1 = df[attribute].quantile(0.05)
    Q3 = df[attribute].quantile(0.95)
    IQR = Q3 - Q1
    outliers = df[(df[attribute] < Q1 - 1.5*IQR) | (df[attribute] > Q3 + 1.5*IQR)]
    if not outliers.empty:
        print(f"Outliers removed for {attribute}:")
        print(outliers)
    df = df[(df[attribute] >= Q1 - 1.5*IQR) & (df[attribute] <= Q3 + 1.5*IQR)]
```

```python
df.shape
```

```python
df.head()
```

```python
# Select the attributes you want to rescale
df_to_scale = df[attributes]

# Instantiate the scaler
scaler = StandardScaler()

# Fit and transform the data
df_scaled = pd.DataFrame(scaler.fit_transform(df_to_scale), columns=attributes)

df_scaled.head()
```

```python
#Finding the Optimal Number of Clusters via elbow
# Elbow-curve/SSD

ssd = []
range_n_clusters = [2, 3, 4, 5, 6, 7, 8]
for num_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=num_clusters, max_iter=50)
    kmeans.fit(df_scaled)
    
    ssd.append(kmeans.inertia_)
    
# plot the SSDs for each n_clusters
plt.plot(ssd)
```

```python
# Silhouette analysis
from sklearn.metrics import silhouette_score
range_n_clusters = [2, 3, 4, 5, 6, 7, 8]

for num_clusters in range_n_clusters:
    
    # intialise kmeans
    kmeans = KMeans(n_clusters=num_clusters, max_iter=50)
    kmeans.fit(df_scaled)
    
    cluster_labels = kmeans.labels_
    
    # silhouette score
    silhouette_avg = silhouette_score(df_scaled, cluster_labels)
    print("For n_clusters={0}, the silhouette score is {1}".format(num_clusters, silhouette_avg))
    
  
```

# automatic recognition of cluster with the lowest overall development index 
I do this by comparing the mean of each cluster for each feature category individually. 

This ranks the cluster for each feature (export, health, imports, income, life_expec, gdpp should be high and child_mort and inflation should be low to be good)

They get points for each ranking, equal to their rank (ranking first = 1 point)

At the end, I sum up the points over all features. The cluster with the highest points wins "worst country cluster"


```python
# Define the columns to include in the calculation
columns = ['child_mort', 'exports', 'health', 'imports', 'income', 'inflation', 'life_expec', 'total_fer', 'gdpp']

# Calculate the mean value of each feature for each cluster
cluster_means = df[columns + ['Cluster_Labels']].groupby('Cluster_Labels').mean()

# Create a dictionary to store the points for each cluster
points = {cluster: 0 for cluster in cluster_means.index}

# Define the features that indicate low development when their values are low
high_value_features = ['exports', 'health', 'imports', 'income', 'life_expec', 'gdpp']

# Define the features that indicate low development when their values are high
low_value_features = ['child_mort', 'inflation', 'total_fer']

# Assign points for low value features
for feature in low_value_features:
    # Rank the clusters by the mean value of the feature
    rankings = cluster_means[feature].rank()
    
    # Assign points based on the rankings
    for cluster, ranking in rankings.items():
        points[cluster] += ranking

# Assign points for high value features
for feature in high_value_features:
    # Rank the clusters by the mean value of the feature in descending order
    rankings = cluster_means[feature].rank(ascending=False)
    
    # Assign points based on the rankings
    for cluster, ranking in rankings.items():
        points[cluster] += ranking

# Find the cluster with the lowest total points
lowest_development_cluster = max(points, key=points.get)

# Display the result
print(f'The cluster with the lowest overall development index is: {lowest_development_cluster}')

# Create a DataFrame from the points dictionary
points_df = pd.DataFrame.from_dict(points, orient='index', columns=['Total Points'])

# Display the result
print(points_df)
```

# critical evaluation

```python
results = []

for k in range(2, 149): #max of 148 individual countries
    # Cutting the Dendrogram based on K
    from scipy.cluster.hierarchy import cut_tree
    cluster_labels = cut_tree(mergings, n_clusters=k).reshape(-1, )
    df['Cluster_Labels'] = cluster_labels

    # Calculate the mean value of each feature for each cluster
    cluster_means = df[columns + ['Cluster_Labels']].groupby('Cluster_Labels').mean()

    # Create a dictionary to store the points for each cluster
    points = {cluster: 0 for cluster in cluster_means.index}

    # Assign points for low value features
    for feature in low_value_features:
        # Rank the clusters by the mean value of the feature
        rankings = cluster_means[feature].rank()
        
        # Assign points based on the rankings
        for cluster, ranking in rankings.items():
            points[cluster] += ranking

    # Assign points for high value features
    for feature in high_value_features:
        # Rank the clusters by the mean value of the feature in descending order
        rankings = cluster_means[feature].rank(ascending=False)
        
        # Assign points based on the rankings
        for cluster, ranking in rankings.items():
            points[cluster] += ranking

    # Find the cluster with the lowest total points
    lowest_development_cluster = max(points, key=points.get)

    # Get the country names from df that belong to this cluster
    countries = df[df['Cluster_Labels'] == lowest_development_cluster]['country'].tolist()

    results.append((k, countries))

# Create a DataFrame from the results list
results_df = pd.DataFrame(results, columns=['K', 'Countries'])

# Display the result
results_df
```
